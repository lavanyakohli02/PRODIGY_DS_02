# Importing Modules 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore") #never display warnings which match
```

## Loading Dataset

```python
titanic_data = pd.read_csv('Titanic_Dataset.csv')
```

## Exploring the Data

```python
df = pd.DataFrame(titanic_data)
df
```

## Handling Missing Data

```python
df.isnull().sum()
```

```python
#filling missing values of 'AGE' column with the median and removing 'Cabin' column due to excess missing values
df['Age'].fillna(df['Age'].median(),inplace = True)
df.drop('Cabin',axis = 1,inplace = True)
```

```python
df.isnull().sum()
```

```python
#removing rows with missing values in specific column('Embarked')
df.dropna(subset = ['Embarked'],inplace = True)
```

```python
df.isnull().sum()
```

## Performing EDA and Data Visualization

```python
#exploring distribution of target variable 'Survived'
sns.countplot(x='Survived', palette=['cyan', 'pink'], edgecolor='black', data=df)
plt.show()
```

```python
#exploring relationship between variables ('Survived' vs 'Sex')
sns.countplot(x = 'Survived', hue = 'Sex', palette = 'Set1', edgecolor = 'black', data = df )
plt.show()
```

```python
sns.catplot(x = 'Survived', col = 'Sex', palette = 'Set1', edgecolor = 'black', kind = 'count', data = df)
plt.show()
```

```python
#checking the exact count of males and females survived
df.groupby(['Sex']).Survived.sum()
```

```python
df['Pclass'] = df['Pclass'].astype(str)
sns.countplot(x = 'Survived', hue = 'Pclass', palette = 'Set1', data = df)
```

```python
sns.violinplot(x = 'Sex', y = 'Age', hue = 'Survived' ,palette = ['red', 'green'], data = df , split = True)
```

```python
sns.countplot(x = 'SibSp', data = df)
```

```python
plt.figure(figsize = (12,7))
sns.boxplot(x = 'Pclass', y = 'Age', palette = 'winter', data = df)
```

```python
#analyze the Age Distribution of Parameters
sns.histplot(df['Age'],bins = 20,color = 'red')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#calculate and visualize the correlation between numerical features
numerical_df = df.select_dtypes(include=['float64', 'int64'])
correlation = numerical_df.corr()
plt.figure(figsize = (15,8))
sns.heatmap(correlation, annot = True, cmap = 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#plottng pairplot
sns.pairplot(df,hue = 'Survived', palette = 'Set2')
plt.show()
```
